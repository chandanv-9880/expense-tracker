```python
class expense:
    
    def __init__(self,name,category,amount)-> None:
        self.name = name
        self.category = category
        self.amount = amount
        now = datetime.datetime.now()
        now = now.strftime("%d-%m-%y %H:%M")
        self.date = now
    
    def __repr__(self):
        return f"<Expense: {self.name}, {self.category}, Rs.{self.amount:.2f}"
```

```python
import datetime
def main():
    print(f"running expense tracker")
    expense_file = 'expense_file.csv'
    #Get user input for expense
    #expense = get_user_expense()
    print(expense)
    #write their expense into a file
    #save_expense_to_file(expense,expense_file)
    
    #Read file and summarize expenses
    summarize_expenses(expense_file)
    
   

def get_user_expense():
    print("getting user expense")
    expense_name = input("Enter expense name:")
    expense_amount = float(input("Enter expense amount:"))
    
    expense_categories = [
        "Food",
        "Home",
        "Travel",
        "Fun",
        "Family",
        "misc"
    ]
    
    while True:
        print("Select a category :")
        for i , category_name in enumerate(expense_categories):
            print(f"  {i+1}.{category_name}")
            
        value_range = f"1 - {len(expense_categories)}"
        try:
            selected_index = int(input(f"Select category number {value_range}:"))-1
            
            if selected_index in range(len(expense_categories)):
                new_expense = expense(name = expense_name,category = expense_categories[selected_index],amount=expense_amount)
                return new_expense
            else:
                print("invalid category number, try again: ")
        except:
            print("invalid input, Enter a number")
        

def save_expense_to_file(expense,expense_file):
    print(f"saving {expense} to file {expense_file}")
    with open(expense_file,'a') as f:
        f.write(f"{expense.name},{expense.category},{expense.amount},{expense.date}\n")

def summarize_expenses(expense_file):
    print("summarizing expenses")
    expenses = []
    with open(expense_file,"r") as f:
        lines = f.readlines()
        for line in lines:
            expense_name,expense_category,expense_amount,expense_date =line.strip().split(",")
            line_expense = expense(name = expense_name,category = expense_category,amount=float(expense_amount))
            expenses.append(line_expense)
           
    amount_by_category = {}
    for exp in expenses:
        key = exp.category
        if key in amount_by_category:
            amount_by_category[key] += exp.amount
        else:
            amount_by_category[key] = exp.amount
            
    print("Expenses by category")
    for key , amount in amount_by_category.items():
        print(f" {key} : Rs{amount:.2f}")
        
    total_spent = sum([ex.amount for ex in expenses])
    print(f"Total spent: {total_spent}")    
    now = datetime.datetime.now()
    now = now.strftime("%d-%m-%y %H:%M")
    total_spent_this_month = sum([ex.amount for ex in expenses if ex.date[3:5] == now[3:5]])
    print(f"Total spent this month : {total_spent_this_month}")
    
if __name__ == "__main__":
          main()
```
